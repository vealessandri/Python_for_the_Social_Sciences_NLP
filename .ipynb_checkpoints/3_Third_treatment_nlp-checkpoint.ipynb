```python
def sentence_to_wordlist(raw):
    clean = re.sub("[^a-zA-Z\u00C0-\u017F]"," ", raw)
    words_from_tweets = clean.split()
    return words_from_tweets
```

```python
sentence_list = [] # lista vazia para receber as frases do corpus e irá 'construir' o vocabulário da vetorização

for sentence in sent_tokenized:
    if len(sentence) > 0:
        sentence_list.append(sentence_to_wordlist(sentence))   
```

```python
tokenizer = MWETokenizer(tweets_text, separator='\n')
```

```python
tweets_text_tokenized = tokenizer.tokenize(tweets_text.split())
```

```python
tweets_text_tokenized
```

```python
# Assim como vetores na física possuem módulo, direção e sentido, 
# também em programação de vetorização de palavras precisamos de parâmetros

# Usaremos distância (posição no plano cartesiano), 
# similaridade (se são semanticamente próximos) 
# e ranking (plot)

# Quanto mais dimensões, mais 'pesado' será para o computador rodar, 
# mas também mais o modelo tenderá à exatidão. Escolhemos um número alto 
# para dimensinalidade (10000) para o número de casos. 

# Dimensionalidade
num_features = 10000

# Parâmetro para o menor tamanho de palavra a ser considerado
min_word_count = 3

# Quanto mais 'threads', mais rapidamente o código é executado (em núcleos paralelos) 
num_workers = multiprocessing.cpu_count() 

# Tamanho da 'janela' de contexto do texto de leitura do texto a ser vetorizado
# Por esse parâmetro, o script 'lê' o texto em blocos de x (7 aqui escolhido como parâmetro) palavras por vez.
context_size = 7

#'Downsample setting' para palavras frequentes, para que a repetição não crie viés na análise
# qualquer ocorrência entre 0-1e-5 perfaz uma medida relevante para quão frequentemente 
# uma mesma palavra repetida será 'lida' pela 'janela de contexto'. Quanto mais frequentemente uma palavra 
# ocorre no texto, tanto mais desejaremos que o algoritmo pondere de modo maleável os vários contextos de 
# ocorrência dessa mesma palavra para gerar o vetor representacional de tal palavra 
downsampling = 1e-3

# Palavras são lidas no contexto de ocorrência pelo algoritmo. A palavra 'foco' é pareada com palavras 
# vizinhas (aqui nesse modelo de 'context-size' igual a 7, o contexto é de três palavras antes e 
# três palavras depois da 'palavra foco'). O parâmetro 'seed' corresponde então a um atributo 
# determinístico aleatório ('RNG' - random number generator) que busca tornar o resultado da análise
# (métrica obtida pelo algoritmo a cada ciclo da rodada de leitura) anterior em parâmetro reutilizável 
# no giro ulterior (epochs) do algoritmo. Esta medida é útil para estipular programaticamente 
# um parâmetro para reutilização de instâncias do corpus de texto para configurar a métrica da
# representação vetorizada. Quanto menor a medida, mais os trechos se tornam reutilizáveis no modelo.
seed = 1
```

```python
tweet2vec = w2v.Word2Vec(
    sg=1,
    seed = seed,
    workers = num_workers,
    size = num_features,
    min_count = min_word_count,
    window = context_size, 
    sample = downsampling
)
```

```python
tweet2vec.build_vocab(sentence_list)
```

```python
print("Quantidade de palavras no vocabulário vetorizado:", len(tweet2vec.wv.vocab))
```

```python
tweet2vec.train(sentence_list, epochs=50, total_examples=tweet2vec.corpus_count)
```

Até aqui, o modelo 'tweets2vec' foi definido com 10000 dimensões. Para ser possível plotar num plano cartesiano, o ideal é reduzir essa multidimensionalidade para duas dimensões (uma dimensão 'x' e uma dimensão 'y').

Para reduzir essa dimensionalidade sem distorcer os dados produzidos, iremos usar o método TSNE "T-Stochastic Neighbor Embedding", uma técnica de redução de multidimensionalidade algoritmo inventado por Josh Barnes e Piet Hut em publicação de 04 de dezembro de 1986 (Nature, 324, 446), artigo intitulado "A hierarchical O(N log N) force-calculation algorithm". Ver: https://www.nature.com/articles/324446a0

##### Para outras referências sobre o algoritmo t-SNE, ver:

L.J.P. van der Maaten and G.E. Hinton. Visualizing Non-Metric Similarities in Multiple Maps. Machine Learning 87(1):33-55, 2012. 
L.J.P. van der Maaten. Learning a Parametric Embedding by Preserving Local Structure. In Proceedings of the Twelfth International Conference on Artificial Intelligence & Statistics (AI-STATS), JMLR W&CP 5:384-391, 2009.  PDF
L.J.P. van der Maaten and G.E. Hinton. Visualizing High-Dimensional Data Using t-SNE. Journal of Machine Learning Research 9(Nov):2579-2605, 2008
https://lvdmaaten.github.io/publications/papers/JMLR_2014.pdf

#### Definição de parâmetros de similitude de palavras


Definição de uma função para, ao passar três parâmetros, o algoritmo responder com um quarto que complementa pares de semelhança medidos pela distância angular (aproximação por similaridade do ângulo cosseno entre vetores tomados dois a dois) entre vetores representacionais das palavras escolhidas 

```python
def similaridade_cos(start1, end1, end2):
    similaridades = tweet2vec.wv.most_similar_cosmul(
        positive = [end2, start1],
        negative = [end1]
    )
    start2 = similaridades[0][0]
    print('''Nos tweets, a palavra \033[1m \"{start1}\"\033[0m está relacionada à palavra \033[1m \"{end1}\"\033[0m, 
assim como a palavra \033[1m \"{end2}\"\033[0m está relacionada à palavra:'''.format(**locals()))
    
    return start2
```

#### Definir subregião do gráfico para aproximação da análise

```python
def plot_region(x_bounds, y_bounds):
    slice = points[(x_bounds[0] <= points.x) & (points.x <= x_bounds[1]) &
        (y_bounds[0] <= points.y) & (points.y <= y_bounds[1])]    

    ax = slice.plot.scatter("x", "y", s=35, figsize=(10,8))
    for i, point in slice.iterrows():
        ax.text(point.x + 0.005, point.y + 0.005, point.palavra, fontsize=11)
```
