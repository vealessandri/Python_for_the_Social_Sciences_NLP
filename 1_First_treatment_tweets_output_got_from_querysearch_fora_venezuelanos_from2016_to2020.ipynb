```python
def multiple_replace(input_text):
    lookup_dict = {'vc':'você', 'vcs': 'vocês','pq': 'por que','ñ': 'não', 
                   'tb': 'também','qntidd': 'quantidade', 'pfv': 'por favor', 
                   'pdte':'presidente', 'srlula':'senhor lula', 'incompetenteos':
                   'incompetente os', 'acontecermaldito': 'acontecer maldito', 
                  'acontecerpresidenta': 'acontecer presidenta'}
    words = input_text.split()
    new_words = []
    for word in words:
        for k,v in lookup_dict.items():
            if word == k:
                word = v
        new_words.append(word)
    return new_words                    
```

#### Criar um dicionário usnado dict .fromkeys() para pegar apenas as palavras não repetidas dos tweets

```python
word_dict_from_tweets = dict.fromkeys(word for word in sorted(multiple_replace(tweets_df_lower['text'].str.cat()))if word.isalpha())
```
