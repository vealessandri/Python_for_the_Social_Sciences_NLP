```python
with open('ciroGomes.txt', 'r', encoding='utf-8') as f:
    content = f.read().lower()
    print(content)
```

```python
sent_tokenizer = nltk.data.load('tokenizers/punkt/portuguese.pickle') # invoca o pacote nltk punkt e convert 
                                                                      # para bit
```

```python
texto_cru = set(sent_tokenizer.tokenize(content)) - set(pt_stopwords)
texto_cru = list(texto_cru)
texto_cru
```

#### Converter frases em lista de palavras, removendo caracteres desnecessarios usando 'regex'

```python
def frases_para_listadepalavras(raw):
    clean = re.sub("[^a-zA-Z\u00C0-\u017F]"," ", raw)
    palavras = clean.split()
    return palavras
```

```python
frases = [] # cria uma lista vazia para receber as frases do corpus. 
            # Será útil para 'construir' o vocabulário da vetorização

for texto in texto_cru:
    if len(texto) > 0:
        frases.append(frases_para_listadepalavras(texto))
```

### Contar a quantidade de tokens

```python
# contagem
token_count = sum([len(frase) for frase in frases])

# print na tela de quantos tokens criados
print("Total de {0} tokens".format(str(token_count)))
```

### Aplicar o algoritmo Word2Vec (vetores são um tipo de "tensor")

<img src="/home/lmw/Downloads/linear-relationships.png">

```python
# Assim como vetores na física possuem módulo, direção e sentido, 
# também em programação de vetorização de palavras precisamos de parâmetros

# Usaremos distância (posição no plano cartesiano), similaridade (se são semanticamente próximos) e ranking (plot)
# Quanto mais dimensões, mais 'pesado' será para o computador rodar, mas também mais generalizado será o modelo

# Dimensionalidade
num_features = 100

# Critério para o menor tamanho de palavra a ser considerado
min_word_count = 3

# Número de linhas de paralelização da execução do código
# Quanto mais 'threads', mais rapidamente o código é executado 
num_workers = multiprocessing.cpu_count() 

# Tamanho da 'janela' de contexto do texto a ser vetorizado (código 'olha' blocos de 7 palavras por vez)
context_size = 7

#'Downsample setting' para palavras frequentes, para que a repetição não crie 'bias' na análise
# qualquer ocorrência entre 0-1e-5 perfaz uma medida relevante para quão frequentemente uma mesma palavra repetida
# será 'lida' pela 'janela' de contexto. Quanto mais frequentemente uma palavra ocorre no texto, menos frequentemente
# desejaremos que o algoritmo percorra essa mesma palavra para gerar vetores representacionais da mesma palavra 
downsampling = 1e-3

# Para o atributo 'RNG' (random number generator) definimos um parâmetro 'seed' para tornar um resultado 
# de ocorrência de palavras vetorizadas (ou de pares de ocorrência) reutilizável em análise ulterior. 
# Esta medida é útil para estipularmos programaticamente que instâncias do corpus iremos usar para transformá-lo em 
# representação vetorizada. É uma medida determinística e não probabilística
seed = 1
```

##### Para referência sobre downsampling 'threshold' ver https://github.com/RaRe-Technologies/gensim/blob/727caf220ba2ebe253beaee73877afbd99b2db02/gensim/models/word2vec.py#L600 

```python
ciro2vec = w2v.Word2Vec(
    sg=1,
    seed = seed,
    workers = num_workers,
    size = num_features,
    min_count = min_word_count,
    window = context_size, 
    sample = downsampling
)
```

```python
ciro2vec.build_vocab(frases)
```

```python
print("Quantidade de palavras no vocabulário vetorizado:", len(ciro2vec.wv.vocab))
```

O algoritmo captura o modelo 'amoedo2vec' com todas as frases em seu corpus e perpassa todo o texto fazendo uma varredura de modo a sintetizar modos como palavras ocorrem em relação ao contexto 

```python
ciro2vec.train(frases, epochs=30, total_examples=ciro2vec.corpus_count)
```

### Compactar os vetores de palavras num plano 2D para plotagem

Para reduzir essa dimensionalidade sem distorcer os dados produzidos, iremos usar o método TSNE "T-Stochastic Neighbor Embedding", uma técnica de redução de multidimensionalidade algoritmo inventado por Josh Barnes e Piet Hut em publicação de 04 de dezembro de 1986 (Nature, 324, 446), artigo intitulado "A hierarchical O(N log N) force-calculation algorithm". Ver: https://www.nature.com/articles/324446a0

#### Para outras referências sobre o algoritmo t-SNE, ver:
L.J.P. van der Maaten and G.E. Hinton. Visualizing Non-Metric Similarities in Multiple Maps. Machine Learning 87(1):33-55, 2012. 
L.J.P. van der Maaten. Learning a Parametric Embedding by Preserving Local Structure. In Proceedings of the Twelfth International Conference on Artificial Intelligence & Statistics (AI-STATS), JMLR W&CP 5:384-391, 2009.  PDF
L.J.P. van der Maaten and G.E. Hinton. Visualizing High-Dimensional Data Using t-SNE. Journal of Machine Learning Research 9(Nov):2579-2605, 2008
https://lvdmaaten.github.io/publications/papers/JMLR_2014.pdf 

<img src="/home/lmw/Downloads/300px-2D_Quad-Tree_partitioning_of_100_bodies.png">

```python
tsne = sklearn.manifold.TSNE(n_components=2, random_state=0)
```

```python
all_word_vectors_matrix = ciro2vec.wv.syn0
```

#### Aplicar o algoritmo t-SNE 

```python
all_word_vectors_matrix_2d = tsne.fit_transform(all_word_vectors_matrix)
```

#### Plotar no plano cartesiano

```python
points = pd.DataFrame(
    [
        (word, coords[0], coords[1])
        for word, coords in [
            (word, all_word_vectors_matrix_2d[ciro2vec.wv.vocab[word].index])
            for word in ciro2vec.wv.vocab
        ]       
    ], 
    columns = ["palavra", "x", "y"])
```

```python
# Coordenadas são geradas pelo algoritmo t-SNE, 
# comprimindo os vetores multidimensionais em duas dimensões, "x" e "y")

points.head(100)
```

 #### Fazendo zoom nas áreas de maior interesse dado pela análise visual

Criamos uma subplot ou subrepresentação do gráfico acima, delimitando coordenadas de x e y 

```python
def plot_region(x_bounds, y_bounds):
    slice = points[(x_bounds[0] <= points.x) & (points.x <= x_bounds[1]) &
        (y_bounds[0] <= points.y) & (points.y <= y_bounds[1])]    

    ax = slice.plot.scatter("x", "y", s=35, figsize=(10,8))
    for i, point in slice.iterrows():
        ax.text(point.x + 0.005, point.y + 0.005, point.palavra, fontsize=11)
```

#### Podemos ainda definir uma função utilizando outras métricas, como distância angular medida pelo cosseno (cosine similariy)

```python
def similaridade_cosseno(start1, end1, end2):
    similaridades = ciro2vec.most_similar_cosmul(
        positive = [end2, start1],
        negative = [end1]
    )
    start2 = similaridades[0][0]
    print("{start1} está relacionado a {end1}, assim como {start2} está relacionado a {end2}".format(**locals()))
    return start2
```

```python
similaridade_cosseno("queremos", "preciso", "mudança")
```
